```python
#Final Correct Code
# AI Disclaimer: Used ChatGPT with prompt: "Explain the recursive memoized approach for counting distinct ways to climb stairs" only to explain the logic.

def climbStairs(n):

  memo = {}

  def climb(n):
    if n == 1:
      return 1
    if n == 2:
      return 2
    if n in memo:
      return memo[n]

    result = climb(n-1) + climb(n-2)
    memo[n] = result
    return result

  return climb(n)
```

##1. Understanding & Breakdown

**Problem:** Given an array (nums) with n objects coloured red, white, or blue, sort them in-place so that the objects of the same colour are adjacent, with the colours in the order red (0), white (1), and blue (2).

**Inputs:** An array (nums) containing integer values (0,1,2) that refer to the colour of the object.

**Outputs:** The sorted array.

**Constraints:**
- n == nums.length
- 1 <= n <= 300
- nums[i] is either 0, 1, or 2

**Plan:**
1. Set up three pointers: low(front), mid(middle), high(end)
2. Use the mid pointer to move through the array while it has not passed high
3. When the current value is 0, place it at the front by swapping it with the element at low, then move low and mid forward
4. When the current value is 1, leave it where it is and move mid forward
5. When the current value is 2, place it at the end by swapping it with the element at high, then move high backward
6. Repeat this process until every element has been sorted

##2. Peer Review & Draft Code

I reviewed the following top solution on LeetCode: https://leetcode.com/problems/sort-colors/solutions/6751648/6-sorting-methods-with-images-cpythonjav-phsy/
- I learned that the Dutch National Flag algorithm can sort an array of 0s, 1s and 2s in one pass efficiently (without using extra space).
- The algorithm uses three pointers (low, mid, and high) to swap elements and sort the array in only one pass.

```python
#Draft Code
def sortColors(self, nums):
  low = 0
  mid = 0
  high = len(nums) - 1

  while mid < high:
    if nums[mid] == 0:
      nums[mid], nums[low] = nums[low], nums[mid]
      low += 1
    elif nums[mid] == 1:
      mid += 1
    else:
        nums[mid], nums[high] = nums[high], nums[mid]
        high -= 1
        return nums
```

##3. Final Correct Code

```python
#Final Correct Code
# AI Disclaimer: Used ChatGPT with Prompt: "Explain the logic of the Dutch National Flag Algorithm in sorting an array" only to explain logic and compare to other sorting algorithms studied in class.

def sortColors(nums):
  low = 0
  mid = 0
  high = len(nums) - 1

  while mid <= high:
    if nums[mid] == 0:
      nums[mid], nums[low] = nums[low], nums[mid]
      low += 1
      mid += 1
    elif nums[mid] == 1:
      mid += 1
    else:
        nums[mid], nums[high] = nums[high], nums[mid]
        high -= 1

  return nums
```
