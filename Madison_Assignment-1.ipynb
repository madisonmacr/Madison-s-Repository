#2. Peer Review and Draft Code

- I learned that I can reverse the integer using a loop with modulo and division rather than converting it to a string

```python
class Solution(object):
    def isPalindrome(self, x: int): -> bool:
        if x < 0:
          return False

        reverse = 0

        while x > 0:
          reverse = reverse * 10 + x % 10
          x = x // 10

          if reverse = x:
            print(True)
          else:
            print(False)

```

#Problem 3: Reverse Integer

#1. Understanding & Breakdown

Problem: Given a signed 32-bit integer (x), reverse its digits. If reversing causes overflow (outside range), return 0.

Input: Integer x
Output: The reversed digit integer, or 0 if overflow

Constraints:
  - 32-bit signed integer range [-2^31,2^31-1]

Plan:
1. Check if integer is outside range: If x is outside 32-bit integer range, return 0
2. Check if integer is negative: If x is negative, make it positive and reverse
3. Initialize a variable to store the reversed number (reverse = 0)
4. Loop through the digits of x
5. For all values of x (x>0):
    - take the last digit (x%10)
    - add it to reverse in the correct position (reverse * 10 + last digit)
    - remove the last digit from x
6. Else, return reversed number


#2. Peer Review & Draft Code

I reviewed the following top solution on LeetCode: https://leetcode.com/problems/reverse-integer/solutions/4055/golfing-in-python/
- I learned that recursive calls can simplify handling negative integers by converting them to positive and applying the same logic
- Using this idea, I wrote a function that handles negative inputs with recursion, extracts each digit, and builds the reversed number

```python
class Solution:
  def reverse(self, x: int) -> int:
    if x < 0:
      return self.reverse(-x)

    reverse = 0

    while x > 0:
      reverse = reverse * 10 + x % 10
      x = x / 10

    return reverse
```

#3. Final Correct Code

```python
#Final working solution
#AI Disclaimer: Used ChatGPT with prompt: "How to handle overflow in Python" to understand 32-bit constraints
##Correct return value, sign handling, and overflow logic were manually implemented

class Solution:
  def reverse(self, x: int) -> int:
    negative = x < 0
    x = abs(x)
    reverse = 0

    while x > 0:
      reverse = reverse * 10 + x % 10
      x = x // 10

    if reverse > 2**31:
      return 0

    return -reverse if negative else reverse
```

#Problem 4: Two Sum

#1. Understand & Breakdown

Problem: Given an array of integers and a target integer, find two distinct numbers in that array whose sum equals the target and return their indices (positions). Assume each input has exactly one solution, and the same element must not be used twice.

Inputs: Array of integers (nums), integer (target)
Output: List of two indices [i,j] such that n[i] + n[j] == target and i != j

Constraints:
  - Only one valid solution per input
  - The same element cannot be used twice
  - Array must have at least 2 elements and at most 10,000 elements
  - Each number (including the target) must be within range: [-10^9,10^9]

Plan:
1. Initialize a hash table (num_map) to keep track of numbers and indices
2. Loop through each number in the array along with its index
3. For the currect number, calculate its complement as `target - num`
4. Check if this complement already exists in the hash table:
  - If it does, return the indices of the complement and th current number
  - If it does not, add the current number and its index to the hash table
5. Repeat until the end of the array. (Since exactly one solution exists, a valid pair is guaranteed)



#2. Peer Review & Draft Code

I reviewed the following top solution on LeetCode: https://leetcode.com/problems/two-sum/solutions/6483290/hash-table-python-c-java-js-c-go-rust-swift-kotlin-php-dart/
- I learned that using a dictionary is more efficient than a list for storing numbers and their indices, because it allows quick lookups for complements
- Using this idea, I wrote a function that iterates through the inputted array, stores each number and its index in a dictionary, and returns the pair of indices when a matching complement is found

```python
class Solution:
  def twoSum(self, nums: List[int], target: int) -> List[int]:
    num_map = []
    for i, num in enumerate(nums):
      complement = target - num
      if complement in num_map:
        return [i, num_map[complement]]
      num_map[num] = i
    return []
```

#3. Final Correct Code

```python
#Final working code
#AI Disclaimer: Used ChatGPT with prompt: "How to efficiently find two indices that sum to target using a hash map" to understand complement logic

class Solution:
  def twoSum(self, nums: List[int], target: int) -> List[int]:
    num_map = {}

    for i, num in enumerate(nums):
      complement = target - num

      if complement in num_map:
        return [num_map[complement], i]

      num_map[num] = i

    return []
```

#Problem 5: Valid Palindrome

#1. Understand & Breakdown

Problem: Given a string, determine if it is a palindrome considering only alphanumeric characters and ignoring cases. Return `True` if the cleaned version of the string reads the same both forwards and backwards.

Input: String s (may include any characters)
Output: Boolean (True and False statements)

Constraints:
  - Non-alphanumeric characters should be ignored (spaces, punctuation)
  - Cases should be ignored ("A" == "a")
  - Length of string must be within range [1,2*10^5]

Plan:
1. Initialize empty list `filtered`
2. For each character c in s:
    - If c is alphanumeric, append its lowercase version to `filtered`
3. Let `r_filtered` be the reverse of `filtered`
4. Return (True or False) whether `filtered == r_filtered`

#2. Peer Review & Draft Code

I reviewed the following top solution of LeetCode: https://leetcode.com/problems/valid-palindrome/solutions/39982/python-in-place-two-pointer-solution/
- I learned that the isalnum() method correctly identifies alphanumeric characters, which allows ignoring spaces, punctuation, and other non-alphanumeric characters
- Using this idea, I wrote a function that loops through the string, filters and lowercases only alphanumeric characters, then compares the filtered string to its reverse to determine if it is a palindrome

```python
class Solution():
  def isPalindrome(self, s: str) -> bool:
    filtered = []

    for c in s:
      if c.isalnum:
        filtered.append(c.lower())

    r_filtered = filtered[::-1]

    if filtered == r_filtered
      return True
    else:
      return False
```

#3. Final Correct Code

```python
#Final working solution
#AI Disclaimer: Used ChatGPT with prompt: "Why is isalnum failing without parentheses and how to correctly reverse a list in Python" only for concept understanding
##Filtering logic, slicing, and return type fixes were done manually

class Solution():
  def isPalindrome(self, s: str) -> bool:
    filtered = []

    for c in s:
      if c.isalnum():
        filtered.append(c.lower())

    r_filtered = filtered[::-1]

    return filtered == r_filtered
```
